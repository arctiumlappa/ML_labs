**Импорты+настройки**

```python
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline  
```

**Создание массива данных+Сохранение в .csv-файл**

```python
def generate_csv():
    x1 = np.linspace(0.01, 10, 1000, endpoint=True).round(2)
    x2 = np.linspace(5, 15, 1000, endpoint=True).round(2)
    y = np.round(5 * np.log(x1) * np.log(2 * x2), 2)
    data = pd.DataFrame({
        'x1': x1,
        'x2': x2,
        'y': y
    })
    data.to_csv('generated_points.csv', index=False)
    return data

data = generate_csv()
data.head()
```

**Графики**

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.grid()
plt.scatter(data['x1'], data['y'], label='y(x1)')
plt.xlabel('x1')
plt.ylabel('y')
plt.legend()

plt.subplot(1, 2, 2)
plt.grid()
plt.scatter(data['x2'], data['y'], label='y(x2)', color='orange')
plt.xlabel('x2')
plt.ylabel('y')
plt.legend()

plt.show()
```

**Статистика**

```python
data.describe().loc[['mean', 'min', 'max']]
```

**Очистка и сохранение**

```python
filtered_data = data[(data['x1'] < data['x1'].mean()) |
                     (data['x2'] < data['x2'].mean())]
filtered_data.to_csv('filtered_data.csv', index=False)
filtered_data.head()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

x1 = data['x1']
x2 = data['x2']
X1, X2 = np.meshgrid(x1, x2)
Y = 5 * np.log(X1) * np.log(2 * X2)

ax.plot_surface(X1, X2, Y, cmap='coolwarm', edgecolor='none')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')

plt.show()
```
