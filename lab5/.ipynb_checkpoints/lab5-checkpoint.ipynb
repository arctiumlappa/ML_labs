1. Произведём масштабирование признаков (scaling).

```python
df = pd.read_csv("sponge.data", header=None, names=['A.CAPAS_DEL_CORTEX', 'A.CAPA_INTERNA_DEL_CORTEX', 'A.CORTEX', 'A.CORTEX_FIBROSO', 'A.CORTEX_SOLO_DE_ESPICULAS_TANGENCIALES', 'A.CUERPOS_EXTRANOS_EN_EL_CORTEX', 'A.GROSOR_DEL_CORTEX', 'A.HACES_DE_ESPICULAS_PRINCIPALES_EN_POMPON_EN_EL_CORTEX', 'A.TILOSTILOS_ADICIONALES_COANOSOMA', 'B.NUMERO_DE_TIPOS_DE_MEGASCLERAS', 'C.TIPO_ESPICULA_PRINCIPAL_DIACTINA_TUBERCULADA', 'C.TIPO_ESPICULA_PRINCIPAL_ESTILO', 'C.TIPO_ESPICULA_PRINCIPAL_ESTILOS_2_TAMANOS', 'C.TIPO_ESPICULA_PRINCIPAL_ESTILO_TILOSTILO', 'C.TIPO_ESPICULA_PRINCIPAL_ESTRONGILOXA', 'C.TIPO_ESPICULA_PRINCIPAL_OXAS', 'C.TIPO_ESPICULA_PRINCIPAL_TILOSTILO', 'D.ESPICULA_PRINCIPAL_ESTILO', 'D.ESPICULA_PRINCIPAL_TILOSTILO', 'D.FORMA_BASE_TILOSTILO_PRINCIPAL', 'E.DISPOSICION_MEGASCLERAS_ECTOSOMICAS_EN_EL_ECTOSOMA', 'E.FORMA_BASE_TILOSTILO_ECTOSOMICO', 'E.FORMA_MEGASCLERA_ECTOSOMICA', 'E.TIPO_MEGASCLERA_ECTOSOMICA', 'F.TIPO_DE_EXOSTILO', 'G.FORMA_MEGASCLERA_INTERMEDIARIA', 'G.TIPO_MEGASCLERA_INTERMEDIARIA', 'H.LONGITUD_MEGASCLERAS', 'I.MICROSCLERAS', 'I.TIPO_MICROSCLERA', 'J.ASTER', 'J.DIAMETRO_ESFERASTER', 'J.TIPO_DE_ASTER', 'J.TIPO_DE_DIPLASTER', 'J.TIPO_DE_ESFERASTER', 'K.FORMA_FINAL', 'L.NUMERO_DE_PAPILAS', 'L.PAPILAS', 'M.COLOR', 'N.SUPERFICIE', 'O.DISPOSICION_ESPICULAR_ESQUELETO', 'P.ALOJA_CANGREJO_ERMITANO', 'P.PERFORANTE', 'P.PSEUDORAICES', 'P.SUSTRATO'], engine='python', na_values='?')

bin_col = ['A.CORTEX', 'A.HACES_DE_ESPICULAS_PRINCIPALES_EN_POMPON_EN_EL_CORTEX', 'C.TIPO_ESPICULA_PRINCIPAL_DIACTINA_TUBERCULADA', 'C.TIPO_ESPICULA_PRINCIPAL_ESTILO', 'C.TIPO_ESPICULA_PRINCIPAL_ESTILOS_2_TAMANOS',  'C.TIPO_ESPICULA_PRINCIPAL_ESTILO_TILOSTILO', 'C.TIPO_ESPICULA_PRINCIPAL_ESTRONGILOXA', 'C.TIPO_ESPICULA_PRINCIPAL_OXAS', 'C.TIPO_ESPICULA_PRINCIPAL_TILOSTILO', 'I.MICROSCLERAS', 'J.ASTER', 'L.PAPILAS', 'P.ALOJA_CANGREJO_ERMITANO', 'P.PERFORANTE', 'P.PSEUDORAICES']
for col in bin_col:
    if col in df.columns:
        df[col] = df[col].map({'NO': 0, 'SI': 1})

cat_col = df.select_dtypes(include=['object']).columns.tolist()
num_col = df.select_dtypes(include=[np.number]).columns.tolist()
for col in cat_col:
    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].mode()[0])

for col in num_col:
    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].median())

label_encoders = {}
for col in cat_col:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    label_encoders[col] = le 

scaler = RobustScaler()
xclas = scaler.fit_transform(df)
```

2. С использованием библиотеки scikit-learn написать программу с использованием алгоритмов кластеризации данных, позволяющую разделить исходную выборку на классы, соответствующие предложенной вариантом задаче.

```python
kmeans = KMeans(n_clusters=12, random_state=42)
kmeans_lbl = kmeans.fit_predict(xclas) 

spctrcls = SpectralClustering(n_clusters=12, random_state=42)
spctrcls_lbl = spctrcls.fit_predict(xclas)

meanshift = MeanShift()
meanshift_lbl = meanshift.fit_predict(xclas)

hdb = HDBSCAN(min_cluster_size=2)
hdb_lbl = hdb.fit_predict(xclas)
```

```python
algorithms = [kmeans, spctrcls, meanshift, hdb]
algorithm_labels = [kmeans_lbl, spctrcls_lbl, meanshift_lbl, hdb_lbl]
algorithm_names = ['K-means', 'Spectral', 'MeanShift', 'HDBSCAN']

data = []
names = []

for i, (algo, algo_lbl) in enumerate(zip(algorithms, algorithm_labels)):
    unique_labels = set(algo_lbl)
    if algorithm_names[i] == 'HDBSCAN':
        unique_labels = set(algo_lbl[algo_lbl != -1])
        
    if len(set(algo_lbl)) <= 1:
        print(f"{algorithm_names[i]} нашли только 1 кластер, метрик нет")
        continue
        
    metrics_dict = {'Silhouette': silhouette_score(xclas, algo_lbl), 'Calinski-Harabasz': calinski_harabasz_score(xclas, algo_lbl),
 'Davies-Bouldin': davies_bouldin_score(xclas, algo_lbl) }
    data.append(metrics_dict)
    names.append(algorithm_names[i])

results = pd.DataFrame(data=data, index=names[:len(data)], columns=['Silhouette', 'Calinski-Harabasz', 'Davies-Bouldin'])

print(results)
```

```python
plt.figure(figsize=(10, 4))
plt.bar(results.index, results['Silhouette'], color='blue', alpha=0.7)
plt.title('Silhouette Score')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.show()

plt.figure(figsize=(10, 4))
plt.bar(results.index, results['Calinski-Harabasz'], color='green', alpha=0.7)
plt.title('Calinski-Harabasz Score')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.show()

plt.figure(figsize=(10, 4))
plt.bar(results.index, results['Davies-Bouldin'], color='red', alpha=0.7)
plt.title('Davies-Bouldin Score')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.show()
```

3. Провести эксперименты и определить наилучший алгоритм кластеризации, параметры алгоритма. Необходимо использовать не менее 3-х алгоритмов. Данные экспериментов необходимо представить в отчете (графики, ход проведения эксперимента, выводы).

```python
# 1 - kmeans
# Ищем оптимальные параметры для kmeans

# Изначальные параметры
kmeans = KMeans(n_clusters=12, random_state=42)
kmeans_lbl = kmeans.fit_predict(xclas) 

# Модифицированные параметры
kmmodo = KMeans(n_clusters=8, random_state=42)
kmmodo_lbl = kmmodo.fit_predict(xclas) 

kmmodd = KMeans(n_clusters=12, random_state=42, init="random")
kmmodd_lbl = kmmodd.fit_predict(xclas) 

kmmodt = KMeans(n_clusters=8, random_state=42, init="random")
kmmodt_lbl = kmmodt.fit_predict(xclas) 

kmmodc = KMeans(n_clusters=12, random_state=42, algorithm="elkan")
kmmodc_lbl = kmmodc.fit_predict(xclas) 

kmmodp = KMeans(n_clusters=8, random_state=42, algorithm="elkan")
kmmodp_lbl = kmmodp.fit_predict(xclas) 

kmmodh = KMeans(n_clusters=12, random_state=42, init="random", algorithm="elkan")
kmmodh_lbl = kmmodh.fit_predict(xclas) 

kmmods = KMeans(n_clusters=8, random_state=42, init="random", algorithm="elkan")
kmmods_lbl = kmmods.fit_predict(xclas) 

# Анализируем
algorithms = [kmeans, kmmodo, kmmodd, kmmodt, kmmodc, kmmodp, kmmods]
algorithm_labels = [kmeans_lbl, kmmodo_lbl, kmmodd_lbl, kmmodt_lbl, kmmodc_lbl, kmmodp_lbl, kmmods_lbl]
algorithm_names = ["kmeans - 0", "kmeans (modd) - 1", "kmeans (modd) - 2", "kmeans (modd) - 3", "kmeans (modd) - 4", "kmeans (modd) - 5", "kmeans (modd) - 6"]
best_scores = [0, 0, 100]
best_algo = "None"

data = []

for i, (algo, algo_lbl) in enumerate(zip(algorithms, algorithm_labels)):
    if len(set(algo_lbl)) <= 1:
        print(f"Вариант kmeans №{i} нашёл только 1 кластер, метрик нет")
        continue
        
    metrics_dict = {'Silhouette': silhouette_score(xclas, algo_lbl), 'Calinski-Harabasz': calinski_harabasz_score(xclas, algo_lbl),
 'Davies-Bouldin': davies_bouldin_score(xclas, algo_lbl) }
    if((metrics_dict['Silhouette'] > best_scores[0]) and (metrics_dict['Calinski-Harabasz'] > best_scores[1]) and (metrics_dict['Davies-Bouldin'] < best_scores[2])):
        best_scores[0] = metrics_dict['Silhouette']
        best_scores[1] = metrics_dict['Calinski-Harabasz']
        best_scores[2] = metrics_dict['Davies-Bouldin']
        best_algo = algorithm_names[i]
    data.append(metrics_dict)

results = pd.DataFrame(data=data, index=algorithm_names[:len(data)], columns=['Silhouette', 'Calinski-Harabasz', 'Davies-Bouldin'])

print(results)
print(f"\nСамый лучший - {best_algo}")
```

```python
plt.figure(figsize=(10, 4))
plt.bar(results.index, results['Silhouette'], color='blue', alpha=0.7)
plt.title('Silhouette Score')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.show()

plt.figure(figsize=(10, 4))
plt.bar(results.index, results['Calinski-Harabasz'], color='green', alpha=0.7)
plt.title('Calinski-Harabasz Score')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.show()

plt.figure(figsize=(10, 4))
plt.bar(results.index, results['Davies-Bouldin'], color='red', alpha=0.7)
plt.title('Davies-Bouldin Score')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.show()
```
